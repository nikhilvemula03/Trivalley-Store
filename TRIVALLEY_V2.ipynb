```python
# Code snippet 0
# Assigning the data url
data_url = 'https://raw.githubusercontent.com/nvamsimohan/DallasDSA/main/Trivalley%20Customer%20Spend%20and%20Buyer%20Flag%20Data.csv'
data_url
```

```python
# Code snippet 1
# Importing pandas package for data processing
import pandas
```

```python
# Code snippet 2
# Reading the data from the web URL
data = pandas.read_csv(data_url)

# Displaying the data
data
```

```python
# Code snippet 3
# Displaying the dataset size/shape
data.shape
```

```python
# Code snippet 4
# Creating a new feature Age based on the BirthDate
import pandas as pd
from datetime import datetime

df = pd.DataFrame(data)

# Convert the BirthDate column to datetime format
df['BirthDate'] = pd.to_datetime(df['BirthDate'])

# Computing the age
current_year = datetime.now().year
df['Age'] = current_year - df['BirthDate'].dt.year

# Adjusting for those who haven't had their birthday yet this year
df.loc[df['BirthDate'].dt.month > datetime.now().month, 'Age'] -= 1
df.loc[(df['BirthDate'].dt.month == datetime.now().month) & (df['BirthDate'].dt.day > datetime.now().day), 'Age'] -= 1

data = df
print(data)
```
